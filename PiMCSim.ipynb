```python
%matplotlib notebook
import numpy as np
import random as r
import matplotlib.pyplot as plt

NbrInside = 0
NbrTotal = 10000000

fig = plt.figure()
ax = fig.add_subplot(111)
circ = plt.Circle((0, 0), radius=1, edgecolor='black', facecolor='None') #draw unit circle
ax.add_patch(circ)
ax.set_facecolor('w')
plt.ion()
plt.axis('square') #make plot output square
plt.xlim([0, 1.1]) #set limits of plot
plt.ylim([0, 1.1])

fig.show()
fig.canvas.draw()

for i in range(0,NbrTotal):
    x = r.random()
    y = r.random()
    point = x**2 + y**2
    if (point) < 1:
        NbrInside += 1
        ax.plot(x, y, 'ro')
    else:
        ax.plot(x,y, 'bo')
    fig.canvas.draw()
    
#for i in range(0,NbrTotal):
#    x = r.random()
#    y = r.random()
#    point = x**2 + y**2
#    if (point) < 1:
#        NbrInside += 1
    
pi = NbrInside/NbrTotal * 4
print(pi)
```
